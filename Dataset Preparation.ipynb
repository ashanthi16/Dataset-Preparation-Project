# **Preparation of raw data**

## ** 1. Loading the Dataset**

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df1 = pd.read_csv("A1_HR_Employee_Data.csv")
df1
```

## ** 2. Check Data Types**

```python
df1.info()
```

Age is the only column showing as incorrect data type and it should be a integer. But not allowing to convert to Integer as it has got some text in a column. 

Age column will be converted to integer after fixing the errors in the column.

## ** 3. Check Typos**

Typo errors will be looked into the columns which have data types as Object and columns with data types integer and float will be checked in for typos when looking for outliers.

### ** Age Column**

```python
df2 = df1.copy()
df2 ["Age"].value_counts()

```

According to the above output, its clearly shown that Age 36a has a typo error and that will be replace by 36 as below.

```python
df3=df2.copy()
df3 ["Age"]= df3 ["Age"].replace('36a', "36")
df3 ["Age"].value_counts()

```

#### ** Converting the Age Column Data Type to Integer**

```python
df3["Age"] = df3["Age"].astype(str).astype(int)
df3.dtypes
```

### ** Resigned Column**

```python
df3 ["Resigned"].value_counts()
```

Replacing Y, NO, N, no to Yes & No accordingly

```python
df3 ["Resigned"] = df3 ["Resigned"].replace (["Y","NO","N","no"],["Yes","No","No","No"])
df3 ["Resigned"].value_counts()
```

### ** Business Travel Column**

```python
df3 ["BusinessTravel"].value_counts()
```

Replacing Travels_Rarely,TRAVEL_RARELY, rarely_travel to Travel_Rarely.

```python
df3 ["BusinessTravel"] = df3 ["BusinessTravel"].replace (["Travels_Rarely","TRAVEL_RARELY","rarely_travel"],"Travel_Rarely")
df3 ["BusinessTravel"].value_counts()
```

### ** Business Unit Column**

```python
df3 ["BusinessUnit"].value_counts()
```

Trying to find the Employee ID which says Female incorrectly in the Bussiness Unit Column and the ID is 9465

```python

find_employee_id  = df3[(df3['BusinessUnit']  == 'Female')]
find_employee_id 
```

It's evident that human error has happened and the employee ID 9465 should have Female as the Gender and Sales as the Business Unit.

Replacing the Female with Sales here.

```python
df3 ["BusinessUnit"] = df3 ["BusinessUnit"].replace ("Female","Sales")
df3 ["BusinessUnit"].value_counts()
```

### ** Gender Column**

```python
df3 ["Gender"].value_counts()
```

Trying to find the Employee ID which says Sales incorrectly in the Gender Column and the ID is 9465

```python
find_employee_id  = df2[(df2['Gender']  == 'Sales')]
find_employee_id 
```

It's evident that human error has happened and the employee ID 9465 should have Female as the Gender and Sales as the Business Unit.

Categories MMale,M,male will be replaced with male and Sales will be replaced with Female 

Extra white space in this column will be taken care seperatly.


```python
df3 ["Gender"] = df3 ["Gender"].replace (["Sales","MMale","M","male"], ["Female","Male","Male","Male"])
df3 ["Gender"].value_counts()
```

### ** Marital Status Column**

```python
df3 ["MaritalStatus"].value_counts()
```

Replacing the Marital Status D to Divorced. Extra white space in this column will be taken care seperatly.

```python
df3 ["MaritalStatus"] = df3 ["MaritalStatus"].replace (["D"],"Divorced")
df3 ["MaritalStatus"].value_counts()
```

### ** Over Time Column**

```python
df3 ["OverTime"].value_counts()
```

No corrections required to this column.

## ** 4. Extra White Spaces**

### ** Marital Status Column**

```python
df3 ["MaritalStatus"].value_counts()
```

Deleting the extra white spaces in this column

```python
df3 ["MaritalStatus"]= df3["MaritalStatus"].str.strip() 
df3 ["MaritalStatus"].value_counts()
```

### ** Gender Column**

```python
df3 ["Gender"].value_counts()
```

Deleting the extra white spaces in this column

```python
df3 ["Gender"]= df3["Gender"].str.strip() 
df3 ["Gender"].value_counts()
```

## ** 5. Upper/Lower - Case**

Applying uppercases to all the text column in the dataframe.

```python
df4 = df3.copy()

cols = ["Resigned","BusinessTravel","BusinessUnit","Gender", "MaritalStatus","OverTime"]

df4 [cols] = df4 [cols].apply(lambda x: x.astype(str).str.upper())
df4 [cols]
```

## ** 6. Sanity Checks**

### ** Checking Duplicate Employee IDs **

```python
df5 = df4.copy()
df5.duplicated(subset = ['EmployeeID'])
```

### ** Checking for Outliers **

Checking for outliers in the columns - data type integer and floaters. 

This will identify the impossible values in each of these columns.

```python
df5 = df4.copy()

outliers = ["Age","EducationLevel","JobSatisfaction","MonthlyIncome","NumCompaniesWorked","PercentSalaryHike","PerformanceRating","AverageWeeklyHoursWorked","TotalWorkingYears","TrainingTimesLastYear","WorkLifeBalance","YearsAtCompany","YearsInRole","YearsSinceLastPromotion","YearsWithCurrManager"]
df5 [outliers].describe()
```

When looking at the above data, its clearly seen that 'Average Weekly Hours Worked' column has a outlier as its maximum hours worked is 400 and that's an impossible number as an employee cannot work 400 hours for a week.

It's evident that entering 400 is a human error and my best guess for this correction is replacing the 400 with 40.

### ** Replacing the 400 with 40 in 'Average Weekly Hours Worked Column'**


```python
df6= df5.copy()

df6 ["AverageWeeklyHoursWorked"] = df6 ["AverageWeeklyHoursWorked"].replace (400,40)

df6 ["AverageWeeklyHoursWorked"].describe()
```

## ** 7. Missing Values**

### **Checking for the Missing Values for Numeric Columns**

```python
df7 = df6.copy()

df7.info()
```

This dataset has 1482 rows and its clearly shown that below columns have missing values as Non- Null count is lower than 1482 in the below columns.

   * Education Level
   * Job Satisfaction
   * Monthly Income
   * Work Life Balance
  

### **Finding the Mean for all columns with Missing Values**

```python
mean_col = ["EducationLevel","JobSatisfaction","MonthlyIncome","WorkLifeBalance"]

df7[mean_col].mean(axis=0)
```

### **Finding the Median for all columns with Missing Values'**

```python
median_col = ["EducationLevel","JobSatisfaction","MonthlyIncome","WorkLifeBalance"]

df7[median_col].median(axis=0)
```

### **Replacing the Missing Values - Numeric Columns**

#### **Monthly Income**

When comparing the both mean and median, I beleive the the best fit  to replace the column with the mean value as there are no impossible outliers.

```python
df8=df7.copy()

df8 ["MonthlyIncome"]= df8 ["MonthlyIncome"].fillna(df8 ["MonthlyIncome"].mean())
```

#### **Education Level / Job Satisfaction / Work Life Balance **

When comparing the both mean and median, I beleive the the best fit is to replace these columns with the median value of each column.

```python

df8 [["EducationLevel","JobSatisfaction", "WorkLifeBalance"]] =df8 [["EducationLevel","JobSatisfaction", "WorkLifeBalance"]].fillna(df8 [["EducationLevel","JobSatisfaction", "WorkLifeBalance"]].median())
```

### **Checking for the Missing Values for Text Columns**

```python
df9 = df8.copy()
pd.set_option("display.max_columns", None)
       
missing_text_data =df9 [(df9 ["Resigned"] == 'NAN')|(df9 ["BusinessTravel"] == 'NAN') |(df9 ["Gender"] == 'NAN')|(df9["MaritalStatus"] == 'NAN')|(df9 ["OverTime"] == 'NAN')] 

missing_text_data
```

Accoording to the above output, values are missing for the below columns

   *Resigned
   
   *Over Time

### **Finding the Mode for Missing Values in Resigned / OverTime Columns**

```python
mode_col = ["Resigned", "OverTime"]

df9[mode_col ].mode(axis=0)
```

I believe the best possible way is to replace the missing text for these two columns is by the most common occuring category for the columns. Therfore replacing the missing values by the mode of the respective columns.

### **Replacing the Missing Values - Resigned / OverTime Columns**

Both columns have got "No' as the mode and replacing NaN with No for both columns.

```python
df10 = df9.copy()
df10 ["Resigned"] = df10 ["Resigned"].replace (["NAN"],"NO")
df10 ["OverTime"] = df10 ["OverTime"].replace (["NAN"],"NO")

```

#### **Checking if all Missing Values Fixed **

According to the below Non-Null Count, its clearly shown that all missing values are fixed and there are no NaN values.

```python
df11 = df10.copy()

df11.info()
```
